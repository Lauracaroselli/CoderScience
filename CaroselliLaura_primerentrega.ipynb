#**Conjunto de datos del Informe Mundial sobre la Felicidad 2023**







##Reseña del Dataset


  El dataset elegido corresponde a los datos obtenidos de la Encuesta Mundial Gallup, utilizados para el Informe Mundial de Felicidad 2023, y fue obtenido desde la página World Happiness Report.
  El dataset corresponde a muestras tomadas desde el año 2005 al 2022.



###Detalle de las columnas

*   Country name: nombre del País
*   Iso alpha: el código de país ISO alfa-3 para el país respectivo
*   Year: año
*   Happiness score: se basa en respuestas a la pregunta principal de evaluación de la vida. La Escalera Cantril pide a los encuestados que piensen en una escalera, en la que la mejor vida posible para ellos es un 10 y la peor vida posible es un 0. Luego se les pide que califiquen sus propias vidas actuales en esa escala de 0 a 10
*   Log GDP per capita: El logaritmo del producto interno bruto (PIB) per cápita del país, una medida de prosperidad económica
*   Social support: una medida del apoyo social percibido y las conexiones sociales disponibles para las personas en el país.
*   Healthy life expectancy at birth: esperanza de vida saludable, número promedio de años que se espera que una persona viva con buena salud en el país.
*   Freedom to make life choices: el grado en que los individuos perciben que tienen la libertad de tomar decisiones y elecciones en la vida.
*   Generosity: el nivel de generosidad o comportamientos altruistas reportados por personas en el país.
*   Perceptions of corruption: corrupción
*   Positive affect: El afecto positivo se define como el promedio de tres medidas de afecto positivo en
GWP: reír, disfrutar y hacer cosas interesantes en la Encuesta Mundial Gallup
ondas 3-7. Estas medidas son las respuestas a las siguientes tres preguntas,
respectivamente: “¿Sonreiste o reíste mucho ayer?” y “¿Experimentaste
los siguientes sentimientos durante GRAN PARTE DEL DÍA de ayer? Qué tal si
¿Disfruto?”, “¿Aprendiste o hiciste algo interesante ayer?”
*   Negative affect: El afecto negativo se define como el promedio de tres medidas de afecto negativo en
PCG. Son preocupación, tristeza y enojo, respectivamente, las respuestas a “¿Lo hiciste?”
¿Experimentaste los siguientes sentimientos durante GRAN PARTE DEL DÍA de ayer?
¿Qué tal la preocupación?”, “¿Experimentaste los siguientes sentimientos durante MUCHO
DEL DÍA de ayer? ¿Qué tal la Tristeza?” y “¿Experimentaste la
siguientes sentimientos durante GRAN PARTE DEL DÍA de ayer? ¿Qué tal la ira?

Filtré por los Nan en la columna "Generosity" y la columna de "Log GDP per cápita" ya que, para tomar dos columnas de ejemplo, éstos valores NaN son insignificantes, pero si retiro las filas Nan de Genorosity por ejemplo, estaría quitando valores del Log GDP per cápita, por lo que no voy a quitar filas en éste momento ni reemplazar los valores NaN por otros valores que le quiten el significado a mi análisis, sinó que serán segregados del estudio a medida que vaya avanzando con los gráficos.

La idea es poder ver cuál es la correlación entre cada variable del dataset individualmente con la felicidad. Previo a eso, realizar análisis de los valores de felicidad a los largo de los años para ver cómo fue su comportamiento.


Con el siguiente gráfico lo que se quiere averiguar es cómo están distribuidos los datos del dataset, se utiliza el Histograma de Matplotlib, con el que se concluye:

*   Que la mayoría de los datos se encuentran distribuidos en valores por encima del 5 y llegan hasta el 8.
*   Que la felicidad se puntuó desde 2, aunque son pocas personas las que puntuan a la felicidad con nota por debajo del 4.
*   El puntaje de felicidad se mantiene centrado entre el 4,5 y el 6,5






```python
fig, ax = plt.subplots (figsize=(10,4))
bines_personalizados = [0,0.5,1, 1.5,2, 2.5,3, 3.5,4, 4.5,5, 5.5,6, 6.5,7, 7.5,8, 8.5,9, 9.5, 10]
ax.hist(x=felicidad["Happiness score"],bins=bines_personalizados)
ax.set_xticks(range(0, 10,1))
ax.set_title("Frecuencia de los valores del Índice de Felicidad")
ax.set_xlabel("Intervalos de Felicidad")
ax.set_ylabel("Valores del Indice de Felicidad")
plt.show()


```

A continuación se buscaba ver los ragos de los quartiles, con el gráfico del Boxplot de Matplotlib, para ver con mayor definición en qué intervalos se distribuían los datos, la conclusión:


*  La mitad de la gente puntúa a la felicidad entre el 2 y el 5,5, y la otra mitad, puntúa a la felicidad entre el 5,5 y el 8
*   Punto medio significativo: El hecho de que la división entre las puntuaciones esté en 5.5, justo en el medio del rango de 0 a 10, sugiere que este punto podría ser significativo en términos de percepción de felicidad. Podría representar un umbral donde las personas experimentan un cambio notable en su nivel de satisfacción o bienestar.
*  Diversidad en la percepción de la felicidad: La distribución de las puntuaciones sugiere una diversidad significativa en la percepción de la felicidad dentro de la población. Mientras que algunos individuos se sienten relativamente menos felices, otros se sienten más felices, lo que refleja una variedad de experiencias y perspectivas.
*  Discrepancia en la satisfacción personal: La diferencia en las puntuaciones podría indicar que hay una discrepancia notable en la satisfacción personal entre diferentes grupos de personas. Aquellos que puntúan más bajo podrían estar experimentando desafíos o dificultades que impactan su felicidad, mientras que aquellos que puntúan más alto podrían estar experimentando un mayor bienestar o satisfacción en sus vidas.





```python
plt.figure(figsize=(10, 6))
plt.boxplot(felicidad['Happiness score'])
ax.set_yticks(range(0, 10,1))
plt.title('Boxplot del Índice de Felicidad')
plt.xlabel('')
plt.ylabel('Índice de Felicidad')
plt.grid(True)
plt.show()
```

El propósito del siguiente gráfico fue visualizar lo visto anteriormente, individualizado por año.

Como conclusión a un análisis de los años pasados en pandemia, se puede decir que la puntuación más alta fue disminuyendo y hubo más amplitud entre el mínimo puntaje con la media. Pero en todos los casos la Mediana se ubica dentro del rango de 5 a 6, a excepción del año 2005 que demostró un intervalo de datos más corto, con los datos más comprimidos y con un mayor nivel de felicidad general.
El nivel más alto de felicidad se mantiene casi estable, mientras que el nivel más bajo de felicidad fluctúa subiendo y bajando cada año confirmano la discrepancia en la satisfacción personal.

Seguidamente se obtiene un DataFrame con las primeras cuatro columnas del dataset inicial, mas la columna del logaritmo de ingresos per cápita, con el fin de quitar las filas con valores NaN y realizar un análisis de la dispersión entre las variables felicidad e ingresos.


```python
felicidad_Log_GDP_per_capita= felicidad[['Country name', 'Iso alpha', 'year', 'Happiness score',
       'Log GDP per capita']]
print(felicidad_Log_GDP_per_capita)
```

```python
print(felicidad_Log_GDP_per_capita.isna().sum()) #para obtener los datos NaN del Dataframe obtenido
```

```python
felicidad_Log_GDP_per_capita.dropna(inplace=True) # para eliminar las filas Nan del dataframe obtenido
felicidad_Log_GDP_per_capita.head()
```

```python
print(felicidad_Log_GDP_per_capita.isna().sum()) #para verificar si quedaron valores NaN
```

Se utilizó el Scatter de la biblioteca Matplotlib, concluyendo que exite una clara relación entre el nivel de felicidad y el ingreso per cápita.

```python
plt.figure(figsize=(8, 6))
plt.scatter(felicidad_Log_GDP_per_capita['Happiness score'], felicidad_Log_GDP_per_capita['Log GDP per capita'], alpha=0.5, color='blue')  # alpha controla la transparencia de los puntos
plt.title('Gráfico de Dispersión entre el ídice de felicidad y el ingreso per cápita')
plt.xlabel('Indice de felicidad')
plt.ylabel('Log Ingreso per cápita')
plt.grid(True)
plt.show()
```

El siguiente, es el mismo gráfico que el anterior, pero se le agregó como tercer variable, los años, para que se pueda visualizar

```python
fig, ax=plt.subplots()
mapeo_colores=ax.scatter(felicidad_Log_GDP_per_capita['Happiness score'], felicidad_Log_GDP_per_capita['Log GDP per capita'], c=felicidad_Log_GDP_per_capita['year'])
fig.colorbar(mapeo_colores)
ax.set_title('Gráfico de Dispersión entre el Índice de Felicidad y el Ingreso Per Cápita')
ax.set_xlabel('Indice de felicidad')
ax.set_ylabel('Log Ingreso per cápita')
plt.grid(True)
plt.show()
```
